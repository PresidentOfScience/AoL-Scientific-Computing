```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# Masukkan data points tahun dan temperatur yang tertera pada soal
year = [1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999]
temperature = [14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438]

# Data points yang diberikan diubah ke dalam bentuk array
x = np.array(year)
y = np.array(temperature)
```

```python
# Plotting Data Points Awal

# Menentukan ukuran grafik plotting
plt.figure(figsize=(8,6))
# Memberi judul, label sumbu-x, dan label sumbu-y
plt.title("Plotting Graph between Year and Temperature")
plt.xlabel("Year (x)")
plt.ylabel("Temperature (y)")
# Mengatur interval nilai yang dimunculkan di sumbu-x
plt.xticks(np.arange(min(x), max(x)+1, 2))
# Menandai titik-titik data sesuai data points pada soal
plt.plot(x, y, "bo")
# Keluarkan plot grafik
plt.show()
```

```python
# Plotting for Least Square Regression

# Menentukan ukuran grafik plotting
plt.figure(figsize=(8,6))
# Memberi judul, label sumbu-x, dan label sumbu-y
plt.title("Relationship Plotting Graph between Year and Temperature using Least Square Regression Method")
plt.xlabel("Year (x)")
plt.ylabel("Temperature (y)")
# Mengatur interval nilai yang dimunculkan di sumbu-x
plt.xticks(np.arange(min(year), max(year)+1, 2))
# Menandai titik-titik data sesuai data points pada soal
plt.plot(x, y, "bo")
# Plot persamaan garis regresi
plt.plot(x, -9.934677576074332 + 0.012155757575916748*x, "r")
# Keluarkan plot grafik
plt.show()
```

```python
# Plotting for Direct Interpolation Method

# Menentukan ukuran grafik plotting
plt.figure(figsize=(10,8))
# Memberi judul, label sumbu-x, dan label sumbu-y
plt.title("Relationship Plotting Graph between Year and Temperature using Direct Interpolation Method")
plt.xlabel("Year (x)")
plt.ylabel("Temperature (y)")
# Mengatur interval nilai yang dimunculkan di sumbu-x
x_interplt = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998])
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.yticks(np.arange(14, 15, 0.1))
# Menandai titik-titik data sesuai data points pada soal
plt.plot(x, y, "bo")

# Plot kurva fungsi interpolasi
# Interpolasi linear
y_interplt = np.array([14.2189, 14.2349, 14.2509, 14.2669, 14.2829, 14.2989, 14.3149, 14.3309, 14.3469])
plt.plot(x_interplt, y_interplt, "-")
# Interpolasi Kuadratik
y_interplt = np.array([14.1550, 14.1475, 14.1468, 14.1527, 14.1654, 14.1847, 14.2107, 14.2434, 14.2827])
plt.plot(x_interplt, y_interplt, "-")
# Interpolasi Kubik
y_interplt = np.array([14.1744, 14.2138, 14.2140, 14.1898, 14.1558, 14.1267, 14.1173, 14.1422, 14.2162])
plt.plot(x_interplt, y_interplt, "-")

# Keluarkan plot grafik
plt.show()
```
