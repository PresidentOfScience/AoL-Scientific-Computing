```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sin(x)

# Inisialisasi titik-titik absis koordinat penyusun kurva
x = np.linspace(-np.pi, np.pi, 200)
# Inisialisasi titik ordinat seluruhnya sebagai 0
y = np.zeros(len(x))

# Memberi label untuk legenda dan mengatur ukuran grafik plotting
labels = ["Suku Pertama", "Suku Kedua", "Suku Ketiga", "Suku Keempat"]
plt.figure(figsize=(10,8))

# Inisialisasi pengiterasi suku deret ke-n mulai dari 0
n = 0

# Lakukan iterasi selama orde deret lebih kecil atau sama dengan 4
while(2*n+1 <= 4):
    # Nilai y ditambah sesuai rumus deret Taylor sin(x) pada n tertentu
    y = y + ((-1)**n*x**(2*n+1))/(np.math.factorial(2*n+1))
    # Keluarkan suku deret dan plot pada grafik
    print(f"Suku ke-{n+1} : {((-1)**n)}x^{2*n+1}/{(2*n+1)}!")
    plt.plot(x, y, label=labels[n])
    # Tambahkan pengiterasi n untuk melanjutkan proses loop
    n += 1

# Plot grafik sin(x) sesungguhnya
plt.plot(x, np.sin(x), "k", label="Analitik")

# Lengkapi kelengkapan grafik lainnya
plt.title("Aproksimasi Deret Taylor sin(x) Orde Empat")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# cos(x)

# Inisialisasi titik-titik absis koordinat penyusun kurva
x = np.linspace(-np.pi, np.pi, 200)
# Inisialisasi titik ordinat seluruhnya sebagai 0
y = np.zeros(len(x))

# Memberi label untuk legenda dan mengatur ukuran grafik plotting
labels = ["Suku Pertama", "Suku Kedua", "Suku Ketiga", "Suku Keempat"]
plt.figure(figsize=(10,8))

# Inisialisasi pengiterasi suku deret ke-n mulai dari 0
n = 0

# Lakukan iterasi selama orde deret lebih kecil atau sama dengan 4
while(2*n <= 4):
    # Nilai y ditambah sesuai rumus deret Taylor cos(x) pada n tertentu
    y = y + ((-1)**n*x**(2*n))/(np.math.factorial(2*n))
    # Keluarkan suku deret dan plot pada grafik
    print(f"Suku ke-{n+1} : {((-1)**n)}x^{2*n}/{(2*n)}!")
    plt.plot(x, y, label=labels[n])
    # Tambahkan pengiterasi n untuk melanjutkan proses loop
    n += 1

# Plot grafik cos(x) sesungguhnya
plt.plot(x, np.cos(x), "k", label="Analitik")

# Lengkapi kelengkapan grafik lainnya
plt.title("Aproksimasi Deret Taylor cos(x) Orde Empat")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# sin(x)cos(x)

# Inisialisasi titik-titik absis koordinat penyusun kurva
x = np.linspace(-np.pi, np.pi, 200)
# Inisialisasi titik ordinat seluruhnya sebagai 0
y = np.zeros(len(x))

# Memberi label untuk legenda dan mengatur ukuran grafik plotting
labels = ["Suku Pertama", "Suku Kedua", "Suku Ketiga", "Suku Keempat"]
plt.figure(figsize=(10,8))

# Inisialisasi pengiterasi suku deret ke-n mulai dari 0
n = 0

# Lakukan iterasi selama orde deret lebih kecil atau sama dengan 4
while(2*n+1 <= 4):
    # Nilai y ditambah sesuai rumus deret Taylor sin(x).cos(x) pada n tertentu
    y = y + ((-1)**n*4**n*x**(2*n+1))/(np.math.factorial(2*n+1))
    # Keluarkan suku deret dan plot pada grafik
    print(f"Suku ke-{n+1} : {((-1)**n)*(4**n)}x^{2*n+1}/{(2*n+1)}!")
    plt.plot(x, y, label=labels[n])
    # Tambahkan pengiterasi n untuk melanjutkan proses loop
    n += 1

# Plot grafik sin(x).cos(x) sesungguhnya
plt.plot(x, np.sin(x)*np.cos(x), "k", label="Analitik")

# Lengkapi kelengkapan grafik lainnya
plt.title("Aproksimasi Deret Taylor sin(x).cos(x) Orde Empat")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
